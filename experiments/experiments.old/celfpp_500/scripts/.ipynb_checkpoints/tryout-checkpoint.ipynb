The structure of a network file for `celfpp` is as follows:
- The first line consists of two numbers: *number of nodes* and *number of edges*
- Every following line has the form: *node1*, *node2*, *probability*
- The node numbers start from 1 and end with *number of nodes*
- There are twice as many edge descriptions as the number of edges. One for the edge from *node1* to *node2*, and one for edge from *node2* to *node 1*.
- It can be that a node has an edge to itself. In that case, the probabilities for both directions are the same.
